### Statistical Tests
Let's perform a t-test to compare the mean age between different groups (PCOS vs. non-PCOS).

```python
# Import necessary library
from scipy.stats import ttest_ind

# t-Test
def t_test(var, target_group):
    group1 = data[data[target_group] == 0][var].dropna()
    group2 = data[data[target_group] == 1][var].dropna()
    
    t_stat, p_value = ttest_ind(group1, group2)
    
    print(f'T-Test for {var} between {target_group} groups')
    print(f'T-Statistic: {t_stat}')
    print(f'P-value: {p_value}')
    
    # Visualization
    plt.figure(figsize=(10, 6))
    sns.histplot(group1, kde=True, label=f'{target_group}=0')
    sns.histplot(group2, kde=True, label=f'{target_group}=1')
    plt.title(f'Distribution of {var} by {target_group}')
    plt.xlabel(var)
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

# Perform t-Test for selected variables
t_test('Age (yrs)', 'PCOS (Y/N)')
t_test('BMI', 'PCOS (Y/N)')
```

# Pairplot

- Scatter Plots: These show the relationship between two numerical variables. Each point on the plot represents a pair of values from the two variables. The pattern of points can indicate the strength and direction of the association.
- Histograms: On the diagonal of the pair plot, you'll see histograms for each variable. These display the distribution of values in that specific variable. It gives an idea of the frequency or density of different values.
- Correlation Coefficients: The numbers displayed on the scatter plots represent the correlation coefficients. A value close to 1 indicates a strong positive correlation, -1 indicates a strong negative correlation, and 0 suggests no linear correlation. It helps understand the degree and direction of the relationship between variables.

# Box Plot
A boxplot is a graphical representation that displays the distribution and spread of a dataset.


1. Box: The box represents the interquartile range (IQR), which is the middle 50% of the data. The bottom and top of the box mark the first (Q1) and third (Q3) quartiles, respectively. The height of the box is the IQR.
2. Line inside the Box (Median): The line inside the box represents the median of the data. It's the point that divides the data into two equal halves.
3. Whiskers: The whiskers extend from the box to the data points outside the box, up to a certain distance. By default, in seaborn, the whiskers extend to 1.5 times the IQR. Points beyond the whiskers are considered outliers.
4. Outliers: Individual data points beyond the whiskers are plotted as individual points. They represent values that are significantly different from the rest of the data.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='PCOS (Y/N)', y='BMI', data=data)
plt.title('BMI Distribution by PCOS Status')
plt.show()
```

# Odds Ratio


The odds ratio is a measure that helps us understand the strength and direction of the association between two variables. In the context of your project, it represents how the odds of an event happening in one group compare to the odds in another group.

For example, if the odds ratio is 1, it suggests that there is no association. A value greater than 1 indicates higher odds in the first group, while a value less than 1 suggests higher odds in the second group.

```python
# Assuming your dataframe is named data
cross_table = pd.crosstab(data['Weight gain(Y/N)'], data['PCOS (Y/N)'])

# Calculate the odds ratio
odds_ratio = cross_table.iloc[1, 1] / cross_table.iloc[0, 1] / (cross_table.iloc[1, 0] / cross_table.iloc[0, 0])

print(f'Odds Ratio: {odds_ratio}')
```

In this specific case, with an odds ratio of 7.421 between PCOS and weight gain, it means that the odds of having weight gain are 7.421 times higher for those who have PCOS as compared to those who don't.

# Adjusted Odds Ratio

```python
import statsmodels.api as sm

model = sm.Logit(data['Weight gain(Y/N)'], sm.add_constant(data[['PCOS (Y/N)', 'Reg.Exercise(Y/N)', 'Fast food (Y/N)']]))
result = model.fit()

# Extract adjusted odds ratios and confidence intervals
adjusted_odds_ratios = result.params[1:]
conf_int = result.conf_int().iloc[1:].values

# Print adjusted odds ratios and confidence intervals
for i, var in enumerate(['PCOS (Y/N)', 'Reg.Exercise(Y/N)', 'Fast food (Y/N)']):
    print(f'Adjusted Odds Ratio for {var}: {adjusted_odds_ratios[i]:.2f}')
    print(f'Confidence Intervals: ({conf_int[i][0]:.2f}, {conf_int[i][1]:.2f})')
    print('\n')
```

1. **PCOS (Y/N):**
   - **Adjusted Odds Ratio (AOR):** 1.68
   - **Confidence Intervals (CI):** (1.25, 2.11)

   Interpretation: Holding constant the effects of regular exercise and fast food consumption, individuals with PCOS are 1.68 times more likely to experience weight gain compared to those without PCOS. The confidence intervals suggest that we are 95% confident that the true adjusted odds ratio lies between 1.25 and 2.11.

2. **Reg.Exercise(Y/N):**
   - **Adjusted Odds Ratio (AOR):** -0.47
   - **Confidence Intervals (CI):** (-0.95, 0.00)

   Interpretation: Holding constant the effects of PCOS and fast food consumption, individuals who engage in regular exercise are associated with lower odds of weight gain. However, it's important to note that the confidence intervals cross 0, indicating that the effect might not be statistically significant.

3. **Fast food (Y/N):**
   - **Adjusted Odds Ratio (AOR):** 1.22
   - **Confidence Intervals (CI):** (0.80, 1.64)

   Interpretation: Holding constant the effects of PCOS and regular exercise, individuals who consume fast food are 1.24 times more likely to experience weight gain. The confidence intervals suggest that we are 95% confident that the true adjusted odds ratio lies between 0.82 and 1.66.

Remember, a confidence interval that includes 1 suggests that the variable may not be a significant predictor of the outcome. In this case, the variable 'Reg.Exercise(Y/N)' seems to have a confidence interval that crosses 0, indicating a potential lack of statistical significance. ## Relative risk
Relative risk (RR) can be calculated using a cross-tabulation (crosstab) by considering the ratio of probabilities or proportions in different groups. 

Our data looks like this:

# Cross-Tabulation Table

|  | Weight gain(Y) | Weight gain(N) |
|------------------|----------------------|----------------------|
| **Fast food(Y)**    | a                   | b                    |
| **Fast food(N)**     | c                    | d                    |


The formula for calculating **Relative Risk (RR)** is:

RR = Probability of outcome in Exposure Group / Probability of outcomein Non-Exposure Group

where:
- P(Exposure ∩ Outcome) is the probability or proportion of the outcome occurring given exposure `a/(a+b)`
- P(No Exposure ∩ Outcome) is the probability or proportion of the outcome occurring given no exposure `c/(c+d)`

```python
# Create a cross-tabulation (crosstab)
cross_table = pd.crosstab(data['Fast food (Y/N)'], data['Weight gain(Y/N)'])

# Sort rows and columns in descending order
cross_table = cross_table.sort_values(by=[1], ascending=False)
cross_table = cross_table.sort_values(by=[1], axis=1, ascending=False)

# Display the cross-tabulation
print(cross_table)
```

```python
# Visualize the pivot table
plt.figure(figsize=(8, 6))
sns.heatmap(cross_table, annot=True, fmt='d', cmap='viridis', cbar=True)
plt.title('Cross-tabulation: Weight gain(Y/N) by Fast food')
plt.show()
```

```python
# Calculate relative risk
exposure_prob = cross_table.loc[1, 1] / cross_table.loc[1].sum()
non_exposure_prob = cross_table.loc[0, 1] / cross_table.loc[0].sum()
relative_risk = exposure_prob / non_exposure_prob

# Display the cross-tabulation and relative risk
print("Cross-tabulation:")
print(cross_table)
print("\nRelative Risk:", relative_risk)
```
